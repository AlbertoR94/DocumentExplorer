```python
import pandas as pd
from langchain_community.document_loaders.csv_loader import CSVLoader

import os
```

```python
WORKING_DIR = os.getcwd()

FILE_NAME = "books.csv"
FILE_PATH = os.path.join(WORKING_DIR, "Data", "BookData", FILE_NAME)
FILE_PATH
```

```python
usecols= ["isbn", "rating", "book_title", "year_of_publication", "Summary"]
df = pd.read_csv(FILE_PATH,  usecols=usecols)
```

```python
df.head()
```

```python
df["description_length"] = df["Summary"].apply(lambda x: len(x))
```

```python
df.head()
```

```python
df.describe()
```

```python
df[df.description_length==1].isbn.nunique()
```

```python
df[df.description_length!=1].isbn.nunique()
```

```python
# We need records that contain the book description, and 
df_with_summary = df[df.description_length!=1]#.drop_duplicates(subset=["isbn"])
```

```python
df_with_summary.describe(percentiles=[0.01, 0.05, 0.10, 0.25, 0.50, 0.75, 0.95])
```

```python
df_with_summary = df_with_summary[df_with_summary.description_length>=40]
```

```python
df_with_summary["year_of_publication"] = df_with_summary["year_of_publication"].astype(int)
```

```python
df_with_summary.to_csv('C:\\Users\\alber\\Documents\\Projects\\BookRecommendation\\Data\\BookData\\books_with_summary.csv', index=False)
```
